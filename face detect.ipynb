```python
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
capture = cv2.VideoCapture(0)
```

```python
c,fullSizeBaseImage = capture.read()
```

```python
gray = cv2.cvtColor(fullSizeBaseImage, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
x = y = 10
```

```python
w = 100 
```

```python
color = (0, 0, 255)
```

```python
cv2.rectangle(fullSizeBaseImage, (x, y), (x + w, y + w), color, 1)
```

```python
plt.imshow(fullSizeBaseImage)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faceRects = classifier.detectMultiScale(
    gray, scaleFactor=1.2, minNeighbors=3, minSize=(32, 32))
```

```python
print(faceRects)
```

```python
if len(faceRects):  # 0
    for faceRect in faceRects:  # 
        x, y, w, h = faceRect
        # 
        cv2.rectangle(fullSizeBaseImage, (x, y), (x + h, y + w), color, 2)
```

```python
plt.imshow(fullSizeBaseImage)
```
